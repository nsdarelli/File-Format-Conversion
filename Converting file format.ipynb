```python
#converting raw data into csv
import csv

raw_data = [
    ["John Doe", 25],
    ["Jane Smith", 30],
    ["Mike Johnson", 35],
    ["Sarah Williams", 28]
]

with open('output.csv', 'w') as f:
  f_write = csv.writer(f)

  f_write.writerow(['Name', 'Age'])

  for row in raw_data:
    f_write.writerow(row)
```

```python
import json

with open('/content/output.csv', 'r') as f:
  f_read = csv.DictReader(f)
  data  = list(f_read)

with open('en.json', 'w') as j:
  json.dump(data, j, indent=4)  #csv to json
```

```python
with open('/content/en.json', 'r') as j:
  json_data = json.load(j)
  headers = list(json_data[0].keys())

with open('ouutput1.csv', 'w') as f:
  f_write1 = csv.DictWriter(f, fieldnames=headers)
  f_write1.writeheader()
  f_write1.writerows(json_data)  #json to csv
```
